```python
import numpy as np 
import matplotlib.pyplot as plt
```

## 3. 

```python
# Declare Variables
k = 2
mu = 0.06
sigma = 0.1
r0 = 0.04
N = 10000
dt = 1 / N
```

```python
rv = np.random.normal(0, 1, N)
process = np.zeros(N)
process[0] = r0

for i in range(1, N):
    
    process[i] = process[i - 1] + (k * (mu - process[i - 1]) * dt) + (sigma * rv[i] * np.sqrt(dt))
```

```python
plt.plot(process)
plt.title("Stochastic Process")
plt.show()
```

```python
integral = np.zeros(N)
integral[0] = r0

# From Problem Two
alpha = mu * (1 - np.exp(k * 1))
print("Alpha: ", alpha)

beta = sigma
print("Beta: ", beta)

for i in range(1, N):
    
    integral[i] = integral[i - 1] + (alpha * dt) + (beta * rv[i] * np.sqrt(dt))
```

```python
plt.plot(integral)
plt.title("Analytical Solution")
plt.show()
```

```python
plt.plot(process)
plt.plot(integral)
plt.title("Compare Solutions")
plt.show()
```

## 4. 

```python
# Generate R1 N times
r1 = np.zeros(N)

for i in range(N):
    r1[i] = r0 + (k * (mu - r0) * dt) + (sigma * rv[i] * np.sqrt(dt))
```

```python
plt.hist(r1, bins = 100, density = True)
plt.title("Process Distribution")
plt.show()
```
