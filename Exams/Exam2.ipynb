```python
import numpy as np
from scipy.stats import norm
from numba import jit
import matplotlib.pyplot as plt
```

```python
def h3(alpha, beta, b, T):
    
    B1 = (beta + alpha) / (beta + 2 * alpha)
    B2 = (2 * alpha) / (beta + 2 * alpha)

    E1 = np.exp(T * (alpha + beta)**2 / 2)
    E2 = np.exp(T * alpha**2 / 2)
    E3 = np.exp(b * (beta + 2 * alpha))

    N1 = norm.cdf(((alpha + beta) * T - b) / np.sqrt(T))
    N2 = norm.cdf(-(alpha * T + b) / np.sqrt(T))
    
    H3 = (B1 * 2 * E1 * N1) + (B2 * E2 * E3 * N2)
    
    return H3
```

# <center> Lookback Call </center> 

```python
def lookback_call(Y, x, r, T, alpha, sigma):
    
    b = np.log(Y / x) / sigma        
    
    H1 = h3(alpha, sigma, b, T)
    H2 = h3(alpha, 0, b, T)
    
    LB = (np.exp(-r * T) * (Y - x)) + np.exp(-(r + alpha**2 / 2) * T) * (Y * H1 - x * H2)
    
    return LB
```

```python
Y = 100
x = 100
T = 1
r = 0.03
sigma = 0.25
alpha = (r - 0.5 * sigma**2) / sigma
call = lookback_call(Y, x, r, T, alpha, sigma)
call
```

# <center> Lookback Put </center>

```python
def lookback_put(Y, x, r, T, alpha, sigma):
    
    b = np.log(Y / x) / sigma        
    
    H1 = h3(alpha, sigma, b, T)
    H2 = h3(alpha, 0, b, T)
    
    LB = (np.exp(-r * T) * (x - Y)) + np.exp((-r + alpha**2 / 2) * T) * (x * H2 - Y * H1)
    
    return LB
```

```python
Y = 100
x = 135
T = 1
r = 0.03
sigma = 0.25
alpha = (r - 0.5 * sigma**2) / sigma
lookback_put(Y, x, r, T, alpha, sigma)
```

# <center> Monte Carlo Solution </center>

```python
@jit(nopython = True)
def lookback_simulation():
    
    count = 1000
    avg = 0
    
    for i in range(count):
        
        S = 100
        K = 135
        r = 0.03
        sigma = 0.25
        
        T = 1
        N = 100
        dt = T / N
        rv = np.random.normal(0, 1, N)
        
        stock_process = np.zeros(N)
        B = np.cumsum(rv * np.sqrt(dt))
        stock_process = S * np.exp((r - 0.5 * sigma**2) * dt + sigma * B)
        
        Y = np.max(stock_process)
        avg += np.maximum(K - Y, 0)
        
    price = avg / count
    
    return price
```

```python
lookback_simulation()
```

# <center> Delta Hedging </center> 

```python
def delta_hedge(V0, stock_process, delta, N):
    
    pi = np.zeros(N)
    pi[0] = V0
    y = pi[0] - (delta[0] * stock_process[0])

    for i in range(1, N):

        pi[i] = (delta[i - 1] * stock_process[i]) + (y * np.exp(r * dt))
        y = pi[i] - (delta[i] * stock_process[i])
        
    return pi
```

```python
S = 100
T = 1
N = 1000
dt = T / N
rv = np.random.normal(0, 1, N)
        
stock_process = np.zeros(N)
B = np.cumsum(rv * np.sqrt(dt))
stock_process = S * np.exp((r - 0.5 * sigma**2) * dt + sigma * B)

Y = 100
r = 0.03 
sigma = 0.25
x = 135
T = np.linspace(0, 1, N)[::-1]
alpha = (r - 0.5 * sigma**2) / sigma 

stock_series = np.zeros(N)
VN = np.zeros(N)
V1 = np.zeros(N)
epsilon = 1e-3

index = 0
for t in T:
    
    stock_series[index] = stock_process[index]
    Y = np.max(stock_series)
    
    VN[index] = lookback_put(Y, x, r, t, alpha, sigma)
    V1[index] = lookback_put(Y + epsilon, x, r, t, alpha, sigma)
    
    index += 1

plt.plot(VN)
plt.plot(V1)
plt.show()
```

```python
delta = (V1 - VN) / epsilon

hedge = delta_hedge(VN[0], stock_process, delta, N)

plt.plot(hedge, label = "Hedge")
plt.plot(VN, label = "Price")
plt.legend(loc = "best")
plt.show()
```
