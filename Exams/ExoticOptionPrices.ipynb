```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

# <center> Delta Hedging with Self Financing Portfolio</center>

```python
def d1(S, K, T, r, sigma):
    
    return (np.log(S / K) + (r + sigma**2 / 2) * T) / (sigma * np.sqrt(T))

def d2(S, K, T, r, sigma):
    
    return d1(S, K, T, r, sigma) - sigma * np.sqrt(T)

def bs_call(S, K, T, r, sigma):
    
    return S * norm.cdf(d1(S, K, T, r, sigma)) - K * np.exp(-r * T) * norm.cdf(d2(S, K, T, r, sigma))

def get_delta(S, K, T, r, sigma):
    
    return norm.cdf(d1(S, K, T, r, sigma))
```

```python
def delta_hedge(V0, stock_process, delta, N):
    
    pi = np.zeros(N)
    pi[0] = V0
    y = pi[0] - (delta[0] * stock_process[0])

    for i in range(1, N):

        pi[i] = (delta[i - 1] * stock_process[i]) + (y * np.exp(r * dt))
        y = pi[i] - (delta[i] * stock_process[i])
        
    return pi
```

```python
S = 100 
K = 100
r = 0.03 
sigma = 0.25

N = 1000
T = 1
dt = T / N

rv = np.random.normal(0, 1, N)
V0 = bs_call(S, K, T, r, sigma)

stock_process = np.zeros(N)
B = np.cumsum(rv * np.sqrt(dt))
stock_process = 100 * np.exp((r - 0.5 * sigma**2) * dt + sigma * B)

price = np.zeros(N)
for i in range(N): 
    price[i] = bs_call(stock_process[i], K, 1 - dt * i, r, sigma)

delta = np.zeros(N)
for i in range(N): 
    delta[i] = get_delta(stock_process[i], K, 1 - dt * i, r, sigma)

plt.plot(delta_hedge(V0, stock_process, delta, N))
plt.plot(price)
plt.show()
```

# <center> Asian Option </center>

## <center> $V = e^{-r  T}\frac{S_{0}}{r T} \cdot (e^{r T} - 1)$ </center>

```python
def asian_option(S, r, T):
    
    return np.exp(-r * T) * (S / r * T) * (np.exp(r * T) - 1)
```

# <center> Asian Option Delta </center>

## <center> $\Delta = \frac{1}{rT} - \frac{e^{rT}}{rT}$ </center>

```python
def asian_delta(r, t):
    
    return (1 / (r * t)) - (np.exp(-r * t) / r * t)
```

# <center> H Functions <center>

## <center> $ 0 \leq b \leq k $ </center>

## <center> $H_{s}(\alpha, 0, k, b) = E[1_{W_{s}\geq k} \cdot e^{\alpha W_{s}}] = e^{s \frac{\alpha^2}{2}} \cdot N(\frac{s\alpha - k}{\sqrt{s}}) $ </center>

```python
def h1(alpha, k, b, T):
    
    return np.exp(T * alpha**2 / 2) * norm.cdf((T * alpha - k) / np.sqrt(T))
```

```python
alpha = 2
k = 2
b = 1
T = 1
h1(alpha, k, b, T)
```

## <center> $k \leq b$ </center> 

## <center> $H_{s}(\alpha, 0, k, b) = e^{s \frac{\alpha^2}{2}} \cdot (N(\frac{s\alpha - b}{\sqrt{s}}) + e^{2\alpha b} \cdot(N(\frac{-s\alpha-b}{\sqrt{s}}) - N(\frac{-s\alpha - 2b + k}{\sqrt{s}})) )$ </center>

```python
def h2(alpha, k, b, T):
    
    E1 = np.exp(T * alpha**2 / 2)
    E2 = np.exp(2 * alpha * b)

    N1 = norm.cdf((T * alpha - b) / np.sqrt(T))
    N2 = norm.cdf((-T * alpha - b) / np.sqrt(T))
    N3 = norm.cdf((-T * alpha - 2 * b + k) / np.sqrt(T))

    H1 = E1 * (N1 + E2 * (N2 - N3))
    
    return H1
```

```python
alpha = 2
k = 1
b = 2
T = 1
h2(alpha, k, b, T)
```

## <center> $ H_{s}(\alpha, \beta, -\infty, b) = \frac{\beta + \alpha}{\beta + 2\alpha} \cdot 2e^{\frac{s(\alpha + \beta)^2}{2}} \cdot N(\frac{(\alpha + \beta)s - b}{\sqrt{s}}) + \frac{2\alpha}{\beta + 2\alpha} e^{\frac{s\alpha^2}{2}} + e^{b(\beta + 2\alpha)} \cdot N(\frac{-\alpha s + b}{\sqrt{s}})$ </center>

```python
def h3(alpha, beta, b, T):
    
    B1 = (beta + alpha) / (beta + 2 * alpha)
    B2 = (2 * alpha) / (beta + 2 * alpha)

    E1 = np.exp(T * (alpha + beta)**2 / 2)
    E2 = np.exp(T * alpha**2 / 2)
    E3 = np.exp(b * (beta + 2 * alpha))

    N1 = norm.cdf(((alpha + beta) * T - b) / np.sqrt(T))
    N2 = norm.cdf(-(alpha * T + b) / np.sqrt(T))
    
    H3 = (B1 * 2 * E1 * N1) + (B2 * E2 * E3 * N2)
    
    return H3
```

```python
alpha = 2
beta = 2
b = 2
T = 1
h3(alpha, beta, b, T)
```

# <center> Lookback Option </center> 

## <center> $ v(t, x, y) = e^{-r(T - t)}y  - x + x e^{-(r + \frac{\alpha^2}{2})(T - t)} H(\alpha, \sigma, -\infty, b) - ye^{-(r + \frac{\alpha^2}{2})(T - t)} H_{T - t}(\alpha, 0, -\infty, b) $ </center> 

```python
def lookback_option(Y, x, r, T, alpha, sigma):
    
    b = np.log(Y / x) / sigma        
    
    H1 = h3(alpha, sigma, b, T)
    H2 = h3(alpha, 0, b, T)
    
    LB = (np.exp(-r * T) * (Y - x)) + np.exp((r + sigma**2 / 2) * T) * (Y * H1 - x * H2)
    
    return LB
```

```python
Y = 100
x = 100
T = 1
r = 0.03
sigma = 0.25
alpha = (r - 0.5 * sigma**2) / sigma
lookback_option(Y, x, r, T, alpha, sigma)
```

# <center> Barrier Option </center>

## <center> $ V_{t} = 1_{max[0, t] S_{t} \le B} [c(t, S_{t}) - w(t, S_{t})]$ </center>

## <center> $ w(t, x) = e^{-(r + \frac{\alpha^2}{2})(T - t)} [x \cdot H(\alpha + \sigma, 0, \frac{log\frac{K}{x}}{\sigma}, \frac{log\frac{b}{x}}{\sigma}) - K \cdot H(\alpha, 0, \frac{log\frac{K}{x}}{\sigma}, \frac{log\frac{b}{x}}{\sigma})]$ </center> 


## <center> $ c(t, x) = x \cdot N(d_{+}) - K e^{-rT} \cdot N(d_{-}) $ </center> 

```python
def barrier_option(S, K, L, r, sigma, alpha, T):
    
    V = 0
    k = np.log(K / S) / sigma
    b = np.log(L / S) / sigma
    
    if k <= b:
        
        H1 = h2(alpha + sigma, k, b, T)
        H2 = h2(alpha, k, b, T)
        
        tempC = bs_call(S, K, T, r, sigma)
        tempV = np.exp(-(r + alpha**2 / 2) * (T)) * (S * H1 - K * H2)    
        V = tempC - tempV
    
    else: 
    
        H1 = h1(alpha + sigma, k, b, T)
        H2 = h1(alpha, k, b, T)
        
        tempC = bs_call(S, K, T, r, sigma)
        tempV = np.exp(-(r + alpha**2 / 2) * (T)) * (S * H1 - K * H2)    
        V = tempC - tempV
    
    return V
```

```python
S = 100
K = 100
L = 110
r = 0.02
sigma = 0.30
alpha = (r - 0.5 * sigma**2) / sigma
T = 1
barrier_option(S, K, L, r, sigma, alpha, T)
```
