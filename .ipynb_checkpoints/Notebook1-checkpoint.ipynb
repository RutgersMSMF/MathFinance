```python
import numpy as np
import seaborn as sns
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import date, datetime
from scipy.stats import skew, kurtosis
from statsmodels.graphics import tsaplots
```

```python
ticker = "VXX"
today = date.today()

prices = yf.download(ticker, 
                   start='2020-01-01', 
                   end=today, 
                   progress=False, auto_adjust=True)

prices.head()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.plot(prices["Close"], label = "Time Series")
ax1.legend(loc = "best")

ax2.plot(np.log(prices["Close"]), label = "Log Time Series")
ax2.legend(loc = "best")
```

```python
log_returns = []
for i in range(len(prices["Close"]) - 1):
    log_returns.append(np.log(prices["Close"][i + 1] / prices["Close"][i]))    
```

```python
fig, (ax1) = plt.subplots(1, 1)

ax1.plot(log_returns, label = "Log Returns")
ax1.legend(loc = "best")
```

```python
fig = tsaplots.plot_acf(log_returns, lags = 20)
plt.show()
```

```python
fig = sns.kdeplot(log_returns)
```

## 5. Binomial Model

#### The binomial model appears to be a robust method for capturing the behavior of a stock time series. 

#### The is a probability that the stock moves higher and there is a probability that the stock moves lower.  

```python
N = 10000 # number of samples
n = 3 # number of steps
dt = 0.1 # step size
sigma = 0.8 # volatility
u = np.exp(sigma*np.sqrt(dt)) # up return
d = 1/u # down return
S0 = 100 # initial stock price
r = 0.01 # interest rate

K = 10 # strike

q = (np.exp(r * dt) - d) / (u - d) # risk neutral probability

print("Up: ", u)
print("Down: ", d)
print("Risk Neutral: ", q)

U = np.random.binomial(n, q, N) # result of flipping a coin n times, with  N samples of size n
S = S0 * u**(U) * d**(n - U)
V = np.exp(-r * dt * n) * np.sum(np.maximum(S - K, 0)) / N

print("Monte Carlo Value: ", V)
```

```python
from py_vollib.black_scholes import black_scholes
```

```python
true = black_scholes('c', S0, K, sigma, r, 3/365)
print("Analytical Value: ", true)
```

```python
((true - V) / V) * 100
```
