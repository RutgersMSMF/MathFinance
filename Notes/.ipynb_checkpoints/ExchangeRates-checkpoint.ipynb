## Geometric Brownian Motion vs Mean Reverting Process

```python
N = 100
T = 1
dt = T / N
mu = 0.05
sigma = 0.20
k = 1
gamma = 1.00

gbm = np.zeros((N, N))
mr = np.zeros((N, N))

for i in range(N):
    
    temp_gbm = np.zeros(N)
    temp_gbm[0] = 100
    
    temp_mr = np.zeros(N)
    temp_mr[0] = 1
    
    rv = np.random.normal(0, 1, N)
    
    for j in range(1, N):
        
        temp_gbm[j] = temp_gbm[j - 1] + (mu * temp_gbm[j - 1] * dt) + (sigma * temp_gbm[j - 1] * rv[j] * np.sqrt(dt))
        temp_mr[j] = temp_mr[j - 1] + (k * (gamma - temp_mr[j - 1]) * dt) + (sigma * temp_mr[j - 1] * rv[j] * np.sqrt(dt))
        
    gbm[i] = temp_gbm
    mr[i] = temp_mr
```

```python
plt.plot(gbm.T)
plt.title("Geometric Brownian Motion")
plt.show()
```

```python
plt.plot(mr.T)
plt.title("Mean Reversion")
plt.show()
```

```python
N = len(eur_usd_hist)

eur_usd_log = np.zeros(N - 1)
eur_usd_prc = np.zeros(N - 1)
brl_usd_log = np.zeros(N - 1)
brl_usd_prc = np.zeros(N - 1)

for i in range(N - 1):
    
    eur_usd_log[i] = np.log(eur_usd_hist["Close"][i + 1] / eur_usd_hist["Close"][i])
    eur_usd_prc[i] = (eur_usd_hist["Close"][i + 1] - eur_usd_hist["Close"][i]) / eur_usd_hist["Close"][i]
    brl_usd_log[i] = np.log(brl_usd_hist["Close"][i + 1] / brl_usd_hist["Close"][i])
    brl_usd_prc[i] = (brl_usd_hist["Close"][i + 1] - brl_usd_hist["Close"][i]) / brl_usd_hist["Close"][i]
                      
eur_usd_log*=100
eur_usd_prc*=100
brl_usd_log*=100
brl_usd_prc*=100
```

```python
plt.plot(eur_usd_log, label = "Log")
plt.plot(eur_usd_prc, label = "prc")
plt.title("Monthly Returns")
plt.show()
```

```python
plt.plot(brl_usd_log, label = "Log")
plt.plot(brl_usd_prc, label = "prc")
plt.title("Monthly Returns")
plt.show()
```
