```python
import numpy as np 
from numba import jit
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
S = 100
r = 0.03 
sigma = 0.25
T = 1
N = 1000
dt = T / N
```

## Vanilla Asian Option

## <center> $ \frac{1}{T} \int_{0}^{T} S_{u} du $ </center>

## <center> $V = e^{-r  T}\frac{S_{0}}{r T} \cdot (e^{r T} - 1)$ </center>

```python
def asian_option(S, r, T):
    
    return np.exp(-r * T) * (S / r * T) * (np.exp(r * T) - 1)
```

```python
price = asian_option(S, r, T)
print("Asian Option Price: ", price)
```

## Vanilla Asian Option Delta

## <center> $\Delta = \frac{1}{rT} - \frac{e^{rT}}{rT}$ </center>

```python
def asian_delta(r, t):
    
    return (1 / (r * t)) - (np.exp(-r * t) / r * t)
```

## Path One

```python
stock_process = np.zeros(N)
rv = np.random.normal(0, 1, N)

B = np.cumsum(rv * np.sqrt(dt))
stock_process = 100 * np.exp((r - 0.5 * sigma**2) * dt + sigma * B)

delta = np.ones(N) * asian_delta(r, T)
price_series = asian_option(stock_process, r, T)

pi1 = np.zeros(N)
pi1[0] = price
print("Initial Price: ", pi1[0])

y = pi1[0] - (delta[0] * stock_process[0])
print("Initial Money Market: ", y)
```

```python
for i in range(1, N):
    
    pi1[i] = (delta[i - 1] * stock_process[i]) + (y * np.exp(r * dt))
    y = pi1[i] - (delta[i] * stock_process[i])
```

```python
plt.plot(pi1, label = "Delta Hedge")
plt.plot(price_series, label = "Option Price")
plt.title("Delta Hedge")
plt.legend(loc = 'best')
plt.show()
```

```python
print(pi1[-1])
print(price_series[-1])
```

## Vanilla Lookback Option



## <center> $ v(t, x, y) = e^{-r(T - t)}y  - x + x e^{-(r + \frac{\alpha^2}{2})(T - t)} H(\alpha, \sigma, -\infty, b) - ye^{-(r + \frac{\alpha^2}{2})(T - t)} H_{T - t}(\alpha, 0, -\infty, b) $ </center> 

## Path One

```python
stock_process = np.zeros(N)
rv = np.random.normal(0, 1, N)

B = np.cumsum(rv * np.sqrt(dt))
stock_process = 100 * np.exp((r - 0.5 * sigma**2) * dt + sigma * B)
maxS = np.max(stock_process)
```

```python
plt.plot(stock_process)
plt.show()
```

```python
r = 0.03
sigma = 0.25
T = np.linspace(1e-5, 1 - 1e-5, N)
K = 100
alpha = (r - 0.5 * sigma**2) / sigma

index = 0
epsilon = 1e-5
V1 = np.zeros(N)
VN = np.zeros(N)
stock_series = np.zeros(N)

for t in T:

    stock_series[index] = stock_process[index]
    
    T = 1 - t
    b = np.log(1) / sigma        

    one_part_first = (sigma + alpha) / (sigma + 2 * alpha) * 2 * np.exp((alpha + sigma) ** 2 * T / 2) * norm.cdf(((alpha + sigma) * T - b) / np.sqrt(T))
    second_part_first = 2 * alpha / (sigma + 2 * alpha) * np.exp(alpha ** 2 * T / 2) * np.exp(b * (sigma + 2 * alpha)) * norm.cdf(-((alpha * T + b) / np.sqrt(T)))

    H_T_first_part = one_part_first + second_part_first

    one_part_second = (0 + alpha) / (0 + 2 * alpha) * 2 * np.exp((alpha + 0) ** 2 * T / 2) * norm.cdf(((alpha + 0) * T - b) / np.sqrt(T))
    second_part_second = 2 * alpha / (0 + 2 * alpha) * np.exp(alpha ** 2 * T / 2) * np.exp(b * (0 + 2 * alpha)) * norm.cdf(-((alpha * T + b) / np.sqrt(T)))

    H_T_second_part = one_part_second + second_part_second

    if index != 0:
        
        S0 = np.max(stock_series)
        K = stock_process[index]
        
        V1[index] = np.exp(-(r + sigma**2 / 2) * T) * ((S0 + epsilon) * H_T_first_part - K * H_T_second_part)
        VN[index] = np.exp(-(r + sigma**2 / 2) * T) * (S0 * H_T_first_part - K * H_T_second_part)
        
    else: 
        V1[index] = np.exp(-(r + sigma**2 / 2) * T) * ((100 + epsilon) * H_T_first_part - 100 * H_T_second_part)
        VN[index] = np.exp(-(r + sigma**2 / 2) * T) * (100 * H_T_first_part - 100 * H_T_second_part)
        
    index+= 1
    
print("Original Option Price: ", VN[0])
```

```python
plt.plot(VN)
plt.title("Lookback Option Price")
plt.show()
```

## Finite Difference Method 

```python
print((V1[0] - VN[0]) / epsilon)
lookback_delta = (V1 - VN) / epsilon
```

## Self Financing Portfolio

```python
pi1 = np.zeros(N)
pi1[0] = VN[0]
print("Initial Price: ", pi1[0])

y = pi1[0] - (lookback_delta[0] * stock_process[0])
print("Initial Money Market: ", y)
```

```python
for i in range(1, N):

    pi1[i] = (lookback_delta[i - 1] * stock_process[i]) + (y * np.exp(r * dt))
    y = pi1[i] - (lookback_delta[i] * stock_process[i])
```

```python
plt.plot(pi1, label = "Delta Hedge")
plt.plot(VN, label = "Option Price")
plt.title("Delta Hedge")
plt.legend(loc = 'best')
plt.show()
```

```python
print(np.max(stock_process) - stock_process[-1])
```

```python
print("Delta Hedge: ", pi1[-1])
print("Option Price: ", VN[-1])
```
