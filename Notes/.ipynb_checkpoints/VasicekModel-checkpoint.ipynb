```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 2 
b = 0.04 
sigma = 0.2

N = 10000
dt = 1 / N
```

```python
vasicek = np.zeros(N)
vasicek[0] = 0
rv = np.random.normal(0, 1, N)

for i in range(1, N):
    
    vasicek[i] = vasicek[i - 1] + (a * (b - vasicek[i - 1]) * dt) + (sigma * rv[i] * np.sqrt(dt))
```

```python
plt.plot(vasicek)
plt.show()
```
