```python
import numpy as np 
from scipy.stats import norm
import matplotlib.pyplot as plt 
```

```python
S = 100 
K = 100
r = 0.03 
sigma = 0.25

N = 1000
T = 1
dt = T / N

rv = np.random.normal(0, 1, N)
```

```python
stock_process = np.zeros(N)

B = np.cumsum(rv * np.sqrt(dt))
stock_process = 100 * np.exp((r - 0.5 * sigma**2) * dt + sigma * B)
```

```python
def d1(S, K, T, r, sigma):
    
    return (np.log(S / K) + (r + sigma**2 / 2) * T) / (sigma * np.sqrt(T))

def d2(S, K, T, r, sigma):
    
    return d1(S, K, T, r, sigma) - sigma * np.sqrt(T)

def bs_call(S, K, T, r, sigma):
    
    return S * norm.cdf(d1(S, K, T, r, sigma)) - K * np.exp(-r * T) * norm.cdf(d2(S, K, T, r, sigma))

def get_delta(S, K, T, r, sigma):
    
    return norm.cdf(d1(S, K, T, r, sigma))
```

```python
price = np.zeros(N)

for i in range(N): 
    
    price[i] = bs_call(stock_process[i], K, 1 - dt * i, r, sigma)
```

```python
delta = np.zeros(N)

for i in range(N): 
    
    delta[i] = get_delta(stock_process[i], K, 1 - dt * i, r, sigma)
```

```python
plt.plot(delta)
plt.title("Delta")
plt.show()
```

## Start at Initial Option Price

#### Compute the price of $V_{0}$ 

```python
pi_process = np.zeros(N)
pi_process[0] = price[0]
print("Initial Price: ", price[0])

y = pi_process[0] - (delta[0] * stock_process[0])
print("Initial Money Market: ", y)
```

```python
for i in range(1, N):
    
    pi_process[i] = (delta[i - 1] * stock_process[i]) + (y * np.exp(r * dt))
    y = pi_process[i] - (delta[i] * stock_process[i])
```

```python
plt.plot(pi_process, label = "Delta Hedge")
plt.plot(price, label = "Option Price")
plt.title("Delta Hedge")
plt.legend(loc = 'best')
plt.show()
```
