```python
def d1(S, K, T, r, sigma):
    
    return (np.log(S / K) + (r + sigma**2 / 2) * T) / (sigma * np.sqrt(T))

def d2(S, K, T, r, sigma):
    
    return d1(S, K, T, r, sigma) - sigma * np.sqrt(T)
```

```python
def bisection_ivol(Price, S, K, T, r):
        
    N = 1000
    tol = 1e-5
    count = 0    
    lower_bound = 0.20 
    upper_bound = 2.00
        
    while(count < N):
        
        sigma = (lower_bound + upper_bound) / 2
        Price_implied = S * norm.cdf(d1(S, K, T, r, sigma)) - K * np.exp(-r * T) * norm.cdf(d2(S, K, T, r, sigma))
        
        # Market Price - Iterative Price
        diff = Price - Price_implied 
        
        if np.abs(diff) < tol: 
            return sigma
        
        # Iterative Price too High
        if diff < 0: 
            upper_bound = (sigma + upper_bound) / 2
        
        # Iterative Price too Low
        if diff > 0: 
            lower_bound = (sigma + lower_bound) / 2
        
        count += 1
    
    return 0
```

```python
C_IV_list = []
C_K_list = []

df_calls = df_calls = call_opt[(call_opt['strike'] >= 150) & (call_opt['strike'] <= 250)]

for i in range(0, len(df_calls)):
    
    K = df_calls.iloc[i, 2]
    Price = (df_calls.iloc[i, 4] + df_calls.iloc[i, 5]) / 2
    CIV = bisection_ivol(Price, meta_last_price, K, ((date_object.date() - today).days / 252), 0.035)
    days_to_exp = (date_object.date() - today).days

    C_IV_list.append(CIV)
    C_K_list.append(K)
```

```python
# Lets Store These Values to Compare Later 
theo = np.zeros(4)
simulated = np.zeros(4)
```

```python
S0 = meta_last_price # Stock Price
ivol = 0.60 # From ATM Ivol Curve

T = ((date_object.date() - today).days / 252)
u = np.exp(ivol * np.sqrt(T))
d = np.exp(-ivol * np.sqrt(T))
r = 0.05
DeltaT = 1
q = (np.exp(r * DeltaT) - d) / (u - d)

#theoretical price of lookback option
n = 1 # number of periods
import itertools
lst = list(itertools.product([0, 1], repeat=n))

price = 0
for w in lst:
    
    S = np.zeros(n + 1)
    S[0] = S0
    
    for i in range(n):
        S[i+1] = S[i] * u**w[i] * d**(1-w[i])
        price += max(S) * q**sum(w) * (1-q)**(n-sum(w))
        
print("The theoretical price is ", np.exp(-r * DeltaT * n) * price)
theo[0] = np.exp(-r * DeltaT * n) * price

# simulated price of lookback option
N = 10000 # number of simulations

price = 0
for i in range(N):
    
    S = np.zeros(n + 1)
    S[0] = S0
    
    for j in range(n):
        
        coin_flip = np.random.uniform(0, 1) < q
        S[j+1] = S[j] * u**coin_flip * d**(1 - coin_flip)
        
    price += max(S)
    
print("The simulated price is ", np.exp(-r * DeltaT * n) * price / N)
simulated[0] = np.exp(-r * DeltaT * n) * price / N
```
