```python
import numpy as np
import yfinance as yf
from dateutil import parser
from scipy.stats import norm
import matplotlib.pyplot as plt 
from datetime import date, datetime
```

## Black Scholes Model

```python
def d1(S, K, T, r, sigma):
    
    return (np.log(S / K) + (r + sigma**2 / 2) * T) / (sigma * np.sqrt(T))

def d2(S, K, T, r, sigma):
    
    return d1(S, K, T, r, sigma) - sigma * np.sqrt(T)
```

```python
def bs_call(S, K, T, r, sigma):
    
    return S * norm.cdf(d1(S, K, T, r, sigma)) - K * np.exp(-r * T) * norm.cdf(d2(S, K, T, r, sigma))
```

```python
def delta(S, K, T, r, sigma):
    
    return norm.cdf(d1(S, K, T, r, sigma))

def theta(S, K, T, r, sigma):
    
    first = -S * norm.pdf(d1(S, K, T, r, sigma)) * sigma / 2 * np.sqrt(T)    
    second = -r * K * np.exp(-r * T) * norm.cdf(d2(S, K, T, r, sigma))
    
    return first + second

def gamma(S, K, T, r, sigma):
    
    return norm.pdf(d1(S, K, T, r, sigma)) / (sigma * np.sqrt(T) * S)
```

## Monte Carlo Simulation

```python
N = 180
dt = 1 / N
rv = np.random.normal(0, 1, N)
```

```python
sigma = 0.3
mu = 0.05
T = N / 365
K = 100

euler = np.zeros(N)
euler[0] = 100

for i in range(1, N):
    
    euler[i] = euler[i - 1] + (mu * euler[i - 1] * dt) + (sigma * euler[i - 1] * rv[i] * np.sqrt(dt))
```

```python
plt.plot(euler)
plt.title("Euler Method")
plt.show()
```

## The "Greeks"

```python
deltas = np.zeros(N)
gammas = np.zeros(N)
thetas = np.zeros(N)

for i in range(N):
    
    t = (N - i) / 365
    deltas[i] = delta(euler[i], K, t, mu, sigma)
    gammas[i] = gamma(euler[i], K, t, mu, sigma)
    thetas[i] = (theta(euler[i], K, t, mu, sigma) / np.sqrt(365))
```

```python
plt.plot(deltas)
plt.title("Delta")
plt.show
```

```python
plt.plot(gammas)
plt.title("Gamma")
plt.show()
```

```python
plt.plot(thetas)
plt.title("Theta")
plt.show()
```

```python
integral = np.zeros(N)
integral[0] = euler[0]

for i in range(1, N):
    
    integral[i] += integral[i - 1] + (thetas[i] * dt) + (deltas[i] * (euler[i] - euler[i - 1])) + (0.5 * gammas[i] * (euler[i] - euler[i - 1])**2)
```

```python
plt.plot(integral, label = "Ito Integral Path")
plt.plot(euler, label = "Euler Method")
plt.legend(loc = 'best')
plt.show()
```

```python
plt.plot(np.abs(euler - integral))
plt.title("Residual")
plt.show()
```
